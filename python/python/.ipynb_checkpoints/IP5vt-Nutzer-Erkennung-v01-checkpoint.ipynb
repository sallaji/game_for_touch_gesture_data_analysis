```python
matches = data['touchDataList']['matched']

print('son mios' + str(len(matches)))
data_to_plot = []
fig, ax = plt.subplots(figsize=(10,10))
ax.set_title('style: {!r}'.format('estilo'), color='C0')
for match in range(0, len(matches)):
    x = []
    y = []
    last_point = 0
    abstand = 0
    for point in range(len(matches[match]['points'])):
        if point != 0:
            last_point = 1
        
        abstand = abstand + ((matches[match]['points'][point]['x'] - matches[match]['points'][point - last_point]['x'])**2 + 
        (matches[match]['points'][point]['y'] - matches[match]['points'][point - last_point]['y'])**2)**0.5
            
        #abstand = ((matches[match]['points'][point + 1]['x'] - matches[match]['points'][point]['x'])**2 + 
        #(matches[match]['points'][point + 1]['y'] - matches[match]['points'][point]['y'])**2)**0.5
        
        y.append(abstand)
        #y.append(abs(screen_size['x'] - matches[match]['points'][point]['x'] - screen_size['y'] - matches[match]['points'][point]['y']))
        timedelta = matches[match]['points'][point]['timestamp'] - matches[match]['points'][0]['timestamp']
        x.append(timedelta)
        print('Match: ' + str(match) + '. Point: ' + str(point) + ' Abstand: ' + str(abstand) + ' Time: '+ str(x[point]))
    r = lambda: random.randint(0,255)
    random_color = '#%02X%02X%02X' % (r(),r(),r())
    match_line = 'Match ' + str(match)
    ax.plot(x, y, random_color , label=match_line)
    data_to_plot.append([x,y])

ax.legend()
```
